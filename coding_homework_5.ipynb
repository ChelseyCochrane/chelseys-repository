```python
# importing relevant modules - specifically, numpy as np, and matplotlib.pyplot as plt

from matplotlib import pyplot as plt

import numpy as np 
```

```python
# importing acc_hits_hw5.csv as a numpy array called acc_hits

url1 = 'https://raw.githubusercontent.com/ChelseyCochrane/chelseys-repository/main/acc_hits_hw5.csv'
acc_hits = np.loadtxt(url1, delimiter = ',')
```

```python
# importing acc_fa_hw5.csv as a numpy array called acc_fa

url2 = 'https://raw.githubusercontent.com/ChelseyCochrane/chelseys-repository/main/acc_fa_hw5.csv'
acc_fa = np.loadtxt(url2, delimiter = ',')
```

```python
# printing shape of acc_hits 

np.shape(acc_hits)
```

```python
# printing shape of acc_fa 

np.shape(acc_fa)
```

```python
# creating a variable called time that has the same length as acc_hits and acc_fa, with equally spaced values between "zero" and "four"

time = np.linspace(0, 4, 8001)
```

```python
# creating an expression to determine the sampling rate of the data using the knowledge that you have 8001 samples, and the total length of each sample is 4 seconds

8001/4
```

```python
# plotting the first trial of acc_hits in one subplot, and the first trial of acc_fa in another subplot. Plot your time variable on the x-axis. Label your axes

fig, ax = plt.subplots(2,figsize=(15,5))

ax[0].plot(time, acc_hits[0])
ax[0].set_xlabel('Time')
ax[0].set_ylabel('Voltage [$\mu V$]')
ax[1].plot(time, acc_fa[0])
ax[1].set_xlabel('Time')
ax[1].set_ylabel('Voltage [$\mu V$]')
```
