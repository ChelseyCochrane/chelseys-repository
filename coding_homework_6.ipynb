```python
# importing lc_hits_hw6.csv as numpy array called lc_hits.

url3 = 'https://raw.githubusercontent.com/ChelseyCochrane/chelseys-repository/main/lc_hits_hw6.csv'
lc_hits = np.loadtxt(url3, delimiter = ',')
```

```python
# importing lc_fa_hw6.csv as numpy array called lc_fa.

url4 = 'https://raw.githubusercontent.com/ChelseyCochrane/chelseys-repository/main/lc_fa_hw6.csv'
lc_fa = np.loadtxt(url4, delimiter = ',')
```

```python
# printing the shape of each array

print(np.shape(acc_hits))
print(np.shape(acc_fa))
print(np.shape(lc_hits))
print(np.shape(lc_fa))
```

```python
# creating a variable called time that has the same length as your four arrays, with equally spaced values between "zero" and "four."

time = np.linspace(0, 4, 8001)
```

```python
#  plot the first trial of acc_hits in red, and the first trial of lc_hits in the same plot in blue in one subplot. In a second subplot, plot the first trial of acc_fa in red, 
# and the first trial of lc_fa in blue. Put time on the x-axis of both of your subplots. Label the axes of your subplots.

fig, ax = plt.subplots(1,2,figsize=(15,5))
ax[0].plot(time, acc_hits[0], color = "red")
ax[0].plot(time, lc_hits[0], color = "blue")
ax[0].set_xlabel('Time')
ax[0].set_ylabel('Voltage [μV]')
ax[1].plot(time, acc_fa[0], color = "red")
ax[1].plot(time, lc_fa[0], color = "blue")
ax[1].set_xlabel('Time')
ax[1].set_ylabel('Voltage [μV]')
```

```python
# computing the coherence for acc_hits and lc_hits
dt = time[1] - time[0] 
K = lc_hits.shape[0]
N = acc_hits.shape[1] 
xf = np.array([np.fft.rfft(x - x.mean()) for x in acc_hits])  
yf = np.array([np.fft.rfft(y - y.mean()) for y in lc_hits])

T = time[-1]
scale = 2 * dt**2 / T
Sxx = scale * (xf * xf.conj()).mean(0)  
Syy = scale * (yf * yf.conj()).mean(0)  
Sxy = scale * (xf * yf.conj()).mean(0)  

cohr = np.abs(Sxy) / (np.sqrt(Sxx) * np.sqrt(Syy))

f = np.fft.rfftfreq(N, dt)                     
plt.plot(f, cohr.real, color = "red")                      
plt.xlim([0, 50])                         
plt.ylim([0, 1])                         
plt.xlabel('Frequency [Hz]')              
plt.ylabel('Coherence')
plt.title('Coherence between acc_hits and lc_hits')
```

```python
# repeating step 10, but with acc_fa and lc_fa. Plot the results from step 10 in red, and the results from this step in blue, in the same plot

dt = time[1] - time[0] 
K = lc_hits.shape[0]
N = acc_hits.shape[1] 
xf = np.array([np.fft.rfft(x - x.mean()) for x in acc_hits])  
yf = np.array([np.fft.rfft(y - y.mean()) for y in lc_hits])

T = time[-1]
scale = 2 * dt**2 / T
Sxx = scale * (xf * xf.conj()).mean(0)  
Syy = scale * (yf * yf.conj()).mean(0)  
Sxy = scale * (xf * yf.conj()).mean(0)  

cohr = np.abs(Sxy) / (np.sqrt(Sxx) * np.sqrt(Syy))

f = np.fft.rfftfreq(N, dt)                     
plt.plot(f, cohr.real, color = "red")                      
plt.xlim([0, 50])                         
plt.ylim([0, 1])                         
plt.xlabel('Frequency [Hz]')              
plt.ylabel('Coherence')
plt.title('Coherence between acc_hits and lc_hits')

dt = time[1] - time[0] 
K = lc_fa.shape[0]
N = acc_fa.shape[1] 
xf = np.array([np.fft.rfft(x - x.mean()) for x in acc_fa])  
yf = np.array([np.fft.rfft(y - y.mean()) for y in lc_fa])

T = time[-1]
scale = 2 * dt**2 / T
Sxx = scale * (xf * xf.conj()).mean(0)  
Syy = scale * (yf * yf.conj()).mean(0)  
Sxy = scale * (xf * yf.conj()).mean(0)  

cohr = np.abs(Sxy) / (np.sqrt(Sxx) * np.sqrt(Syy))

f = np.fft.rfftfreq(N, dt)                     
plt.plot(f, cohr.real, color = "blue")                      
plt.xlim([0, 50])                         
plt.ylim([0, 1])                         
plt.xlabel('Frequency [Hz]')              
plt.ylabel('Coherence')
plt.title('Coherence between hits and false alarms')
```
